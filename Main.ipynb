```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arquivo = 'basededados.xlsx'
df = pd.read_excel(arquivo)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['PIB percapita'] = (df['PIB (Bilhões de USD)'] * 1000000) / df['Pop']
```

```python
df.groupby(df['Pais'])['PIB (Bilhões de USD)'].sum().plot.barh(title='Maiores PIB', figsize=(8,6));
plt.xlabel('PIB (Bilhões de USD)');
plt.ylabel('Pais');
plt.legend();
```

```python
df.groupby(df['Pais'])['PIB percapita'].sum().plot.barh(title='PIB percapita', figsize=(8,6))
plt.xlabel('PIB percapita');
plt.ylabel('Pais');
plt.legend();
```

```python
df.groupby(df['Pais'])['Área(km2)'].sum().plot.barh(title='Área territorial', figsize=(8,6));
plt.xlabel('Área');
plt.ylabel('Pais');
plt.legend();
```

```python
df.groupby(df['Pais'])['hab/km2'].sum().plot.barh(title='Habitantes por Km2', figsize=(8,6));
plt.xlabel('hab/km2')
plt.ylabel('Pais');
plt.legend();
```
